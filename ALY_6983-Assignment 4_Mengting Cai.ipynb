### GRADED : Boston housing prices
We will use the Boston house pricing dataset to predict house prices based on properties of the house

```python
import numpy as np
import pandas as pd
import sklearn
from sklearn import ensemble
from sklearn import datasets
from sklearn import model_selection
from sklearn import linear_model
import statsmodels.formula.api as smf
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt

boston = sklearn.datasets.load_boston()
print(boston.DESCR)
```

```python
# Split the data into train and test set
X = boston.data
y = boston.target
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.3)
```

```python
#GRADED
# (a) Use statsmodels to build a linear regression model with X_Train, y_train as training data
results = smf.ols('y_train~X_train',data=boston).fit()
print(results.params)
results.summary()
```

```python
# (b) Generate predictions on X_test and call it yhat
# model = smf.OLS(...)
# result = model.fit()
# yhat = result.predict(...)
# Measure the mean square error (https://en.wikipedia.org/wiki/Mean_squared_error)
# mse_error = np.mean((yhat-y_test) ** 2)
from patsy import dmatrices
y,X = dmatrices('y_test ~ X_test', data=boston, return_type='dataframe')
print("X:",type(X))
print(X.columns)
model=smf.OLS(y, X)
result=model.fit()
result.summary()
yhat = result.predict()
mse_error = np.mean((yhat - y_test)**2)
#(c) Print the mean squared error
print("mse_error:",mse_error)
```

```python
# (d) Now, build the same model using sklearn.ensemble.RandomForestRegressor() while using X_train, y_train as training data
# (e) Generate mean square error on the test set
model2 = sklearn.ensemble.RandomForestRegressor()
result2 = model2.fit(X,y)
yhat2 = result2.predict(X)
mse_error2=np.mean((yhat2-y_test)**2)
print(mse_error2)
```

```python
#(f) Compare the mean square error and determine which regressor is better
print("mean squared error of statsmodels", mse_error)
print("mean squared error of sklearn", mse_error2)
if mse_error < mse_error2:
    print("statsmodels is better")
else:
    print("sklearn is better")
```
